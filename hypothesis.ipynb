```python

import pandas as pd
import numpy as np
import scipy
from scipy import stats
```

```python
data= pd.read_csv(r'C:\Users\HP\Downloads\Sales_add (1).csv')
data
```

# The company wishes to clarify whether there is any increase in sales after stepping into digital marketing.

```python
des = data.describe()
des
```

```python
#finding t test & p value using ttest
#t-test can be done using the scipy function stats.ttest_rel()

ttest,pval = stats.ttest_rel(data['Sales_before_digital_add(in $)'], data['Sales_After_digital_add(in $)'])
print('t test value is :', ttest)
print('p value is :',pval)
```

```python
# calculating the critical value
# alpha = 0.05
# degree of freedom df = 22-1
# critical_value= scipy.statst.ppf(1.0 - alpha, df)

cmp =  scipy.stats.t.ppf(q=1-0.05,df=21)
cmp
```

```python
if abs(ttest) > abs(cmp) and pval < 0.05:
    print("Reject the Null Hypothesis H0")
else:
    print("Reject the Alternative Hypothesis H1")
```

# ● The company needs to check whether there is any dependency between the features “Region” and “Manager

```python
datac = data.groupby('Region')['Manager'].value_counts().unstack()
data
```

```python
# chi-squared test can be calculated in Python using the chi2_contingency() SciPy function.
#chi2: The test statistic
#p: The p-value of the test
#dof: Degrees of freedom
#expected: The expected frequencies, based on the marginal sums of the table

chi2,p,dof,expected = stats.chi2_contingency(datac)
```

```python
chi2_critical = stats.chi2.ppf(1-0.05,4)
chi2_critical
```

```python
if chi2 > chi2_critical and p < 0.05 :
    print("Reject Null Hypothesis H0")
else:
    print("Reject the Alternative Hypothesis H1")
```

Conclusion: Finding that the variables are independent (fail to reject H0). So we can conclude that there is no dependency between the features “Region” and “Manager”.
